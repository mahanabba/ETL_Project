```python
import pandas as pd
import numpy as np
```

```python
nfl_df = pd.read_csv("Resources/nfl_draft.csv")
penalties_df = pd.read_csv("Resources/players_penalties.csv")
```

```python
list(nfl_df)
```

```python
penalties_df
```

```python
merge_df = pd.merge(nfl_df, penalties_df, on='Player', how="inner")
```

```python
merge_df = merge_df.drop(labels=["Biography", "Wikipedia Profile", "Image", "Unnamed: 32"], axis=1)
```

```python
merge_df["Fined Amount"] = merge_df["Fined Amount"].str.replace(',','')
merge_df["Fined Amount"] = merge_df["Fined Amount"].str.replace('$','')
merge_df["Fined Amount"] = merge_df["Fined Amount"].astype(int)
```

```python
groupbyyear_df = merge_df.groupby("Year")
groupbyyear_df["Fined Amount"].sum()
```

```python
groupbycollege = merge_df.groupby("College/Univ")
groupbycollege.count
```

```python
groupbycollege["Fined Amount"].sum().sort_values(ascending=False)
```

```python
merge2_df = merge_df.copy(deep=True)
merge2_df["Age first fine (approximate)"] = merge2_df["Age first fine (approximate)"].round()
```

```python
groupedage = merge2_df.groupby("Age first fine (approximate)")
groupedage.count()
```

```python
groupedbypos = merge_df.groupby("Pos")
groupedbypos.count()
```
